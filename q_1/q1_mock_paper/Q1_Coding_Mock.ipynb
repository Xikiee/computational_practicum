## Part 1: Python basics with Van der Waals equation

The Van der Waals equation is given, which is an equation of state that adjusts the ideal gas law to consider the volume of gas molecules and the intermolecular forces in real gases:
\begin{equation}
\tag{Eq. 1}
\left(P+a\left(\frac{n}{V}\right)^2\right)(V-nb)=nRT
\end{equation}
where:

- $a = \frac{27}{64}\frac{R^2 T_C^2}{P_C} = 9.378\cdot10^{-4} \: \frac{m^6 kPa} {mol^2}$
- $b = \frac{1}{8}\frac{R T_C}{P_C} = 9.035\cdot10^{-5} \: \frac{m^3} {mol}$
- V is the volume of the system
- n is the number of moles
- T is the temperature
- R is the universal gas constant
- $T_C$ and $P_C$ are the critical temperature and pressure, respectively.

The thermodynamic parameters of the problem are the following:

$R = 8.314\cdot10^{-3} \: \frac{m^3 kPa} {mol K}$

$T_C = 369.9 \: K$

$P_C = 4254.6 \: kPa$

### Ex 1.1: Python Variables

### Ex 1.2: Python Functions
Define a python function called `van_der_waals`, taking as input 4 arguments (moles(numpy array), temperature (float), pressure(float), volume(float)) in the exact order. The aim is to describe the mathematical function:
\begin{equation}
\tag{Eq. 2}
f(n, T, P, V) = \left(P+a\left(\frac{n}{V}\right)^2\right)(V-nb)-nRT
\end{equation}

```python
def van_der_waals(n:np.ndarray,T:float,P:float,V:float)->np.ndarray:
    """This function solves for the residual of the Van der Walls eqution
    Input: 
    n           : np.ndarray
        moles of the compound
    T     : float
        temperature 
    P        : float
        pressure
    V          : float
        volume
    
    Return:
    Residual        : float
    """
    a = 9.378e-4
    b = 9.035e-5
    R = 8.3145e-3

    return (P+a*(n/V)**2)*(V-n*b) -n*R*T
```

## Part 3: Numerical integration with PFR design equation

Given the first-order elementary reaction in isothermal conditions:
\begin{equation}
\tag{Eq. 3}
A \xrightarrow{\text{k}} B
\end{equation}


You can use the design equation for a PFR to estimate the reactor volume needed to achieve a certain conversion $\chi$ of the reactant A.
\begin{equation}
\tag{Eq. 4}
V = F_A^{in} \displaystyle \int_{0}^{\chi_{target}}\frac{1}{-R_A} d\chi
\end{equation}

where:

$R_A = -kC_A \: [mol/(m^3 \cdot s)]$

Considering:
- $F_A^{in} = 2 \: mol/s$
- $k = 0.5 \: s^{-1}$
- $C_A^{in}= 1 \: mol/m^3$

And using the definition of conversion:

\begin{equation}
\tag{Eq. 5}
V = \frac{F_A^{in}}{C_A^{in}\cdot k} \displaystyle \int_{0}^{\chi_{target}}\frac{1}{(1-\chi)} d\chi
\end{equation}

### Ex. 3.1: Composite trapezoid method for numerical integration
By using the **composite trapezoid method** for numerical integration, find the volume necessary to achieve $\chi_{target} = 0.5$ conversion of A, by using $(Eq. 5)$ and the given data. 

**Ex. 3.1.1**
Create a Python function implementing a generic composite trapezoid method for numerical integration, having the following arguments:
  - A callable function to be integrated
  - The two domain boundaries
  - The number of points for the domain discretization

```python
def composite_trapezoid(func:Callable, a:float, b:float, N: int)->float:
    """Integrates a function 'func' using the composite trapezoid method
    Input:
    func            : Callable
        Function which is to be integrated 
    a               : float
        Initial boundary condition 
    b               : float
        Final boundary condition 
    N               : int 
        Number of points for integration

    Returns: 
    integration     : float 
    """
    x = np.linspace(a,b,N)
    h = (b-a)/(N-1)

    integration = 0
    for i in range(1,N):
        integration += h/2 * (func(x[i]) + func(x[i-1]))
    return integration
```

**Ex. 3.1.2**

**Note** here the conversion $\chi$ refers to the **decimal** conversion as opposed to the percentage conversion.

```python
def integrand(chi: float) -> float:
    """Function returning the value of the integrand function

    Args:
        chi (float): Conversion, decimal.

    Returns:
        float: Integrand function value
    """
    f = 1/(1-chi)
    return f
```

### Ex. 4.1: 1D boundary value problem with finite difference method (FDM)
We aim at solving the following second-order non-autonomous differential equation on the interval $0 \leq x \leq \pi/2$. Our objective is to numerically approximate the function $y(x)$ that is the solution of the following problem.

\begin{equation}
\tag{Eq.11}
\cfrac{\text{d}^2y}{\text{d}x^2} + x^2y = 0
\end{equation}

Find the approximated function $y(x)$ that is the solution of the differential equation written above, supplied with the following Dirichlet boundary conditions:

\begin{equation}
\tag{Eq.12}
 y(x=0) = 1  \text{          and         }  y(x=\pi/2)=1
\end{equation}

Please, complete the following steps:

**Ex. 4.1.1**

Define a Python function called `discretization` that is returning the FDM matrix A and vector b for the above differential equation.
- `discretization` has three arguments: left domain boundary, right domain boundary and number of intervals
- Use the central differencing scheme for the second order derivative

```python
def discretization(
    x_l: float, x_r: float, num_intervals: int
) -> Tuple[np.ndarray, np.ndarray]:
    """A function implementing the finite difference method (FDM) for discretizing a boundary value problem (BVP).
    2nd Order derivative discretized with central difference scheme(CDS)

    Args:
        x_l (float): Left domain boundary
        x_r (float): Right domain boundary
        num_intervals (int): number of intervals in the discretization grid

    Returns:
        tuple(np.ndarray): Matrix A and vector b for FDM
    """
    x= np.linspace(x_l, x_r, num_intervals)
    h = x[1] - x[0] 

    matrix_size = num_intervals-2
    A = np.eye(matrix_size) * (-2 + x[1:-1]**2*h**2)
    A += np.diag(np.ones(matrix_size-1),k=-1)
    A += np.diag(np.ones(matrix_size-1),k=+1)

    # solve the b matrix: 
    b =np.zeros(matrix_size)
    b[0] = -1            # y(x=0) =1 dirichlet boundary condition  
    b[-1] = -1           # y(x=pi/2) = 1 the second dirichlet boundary condition
    return A, b
```

**Ex. 4.1.2**

Define a function called `solve_BVP` which make use of `discretization` and `numpy.linalg.solve` to solve the given BVP.

```python
def solve_BVP(x_l: float, x_r: float, num_intervals: int) -> np.ndarray:
    """Solve a boundary value problem after discretizing with Finite-Difference Method

    Args:
        x_l (float): The left Dirichlet boundary condition
        x_r (float): The right Dirichlet boundary condition
        num_intervals (int): The number of discretization intervals

    Returns:
        np.ndarray: The solution vector
    """
    # unpack the disretization:
    A, b = discretization(x_l,x_r,num_intervals)
    sol = np.zeros(num_intervals)
    sol[0] = 1
    sol[1:-1] = np.linalg.solve(A,b)
    sol[-1] = 1
    return sol

x0 = 0
x1 = np.pi/2
n_points = 100
solution = solve_BVP(x0,x1,n_points)
print(solution)
```
